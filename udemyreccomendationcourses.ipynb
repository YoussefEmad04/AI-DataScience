```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
cv_m=cv.fit_transform(df['clean_title'])
cv_m.toarray()
```

# cosine similarity build RS

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim_m = cosine_similarity(cv_m) #cosine similarity matrix
cos_sim_m
```

```python
course_index=pd.Series(df.index,index=df['course_title']).drop_duplicates()
course_index
```

```python
cos_sim_m[index]
```

```python
score=list(enumerate(cos_sim_m[index])) # here will find similar objects based on the title i sendd to him
```

```python
sorted_score=sorted(score,key=lambda x:x[1],reverse=True)
sorted_score
```

```python
sorted_indices=[i[0]for i in sorted_score[1:]]
sorted_indices
```

```python
sorted_values=[i[1]for i in sorted_score[1:]]
sorted_values
```

```python
recomed_result_df=df.iloc[sorted_indices]
recomed_result_df
```

```python
recomed_result_df['similarity']=np.array(sorted_values)
recomed_result_df
```

# now we will for loop on all courses

```python
def recommend_course(title,numerc=10):
  course_index=pd.Series(df.index,index=df['course_title']).drop_duplicates()
  index=course_index[title]
  score=list(enumerate(cos_sim_m[index]))
  sorted_score=sorted(score,key=lambda x:x[1],reverse=True)
  selected_course_index=[i[0]for i in sorted_score[1:]]
  selected_course_score=[i[1]for i in sorted_score[1:]]
  result_df=df.iloc[selected_course_index]
  result_df['similarity']=selected_course_score
  final_rec=result_df[['course_title','similarity','url','price','num_subscribers']]
  return final_rec.head(numerc)
```

```python
recommend_course('Beginner to Pro - Financial Analysis in Excel 2017',10)
```
