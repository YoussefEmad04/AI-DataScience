### Importing Dataset

```python
y_pred = classifier.predict(X_test) 
y_pred
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True, fmt="d")
```

## K-NN | Full Case-study (Social Media Advertising)

### Importing Dataset & Extracting Features

```python
ads = pd.read_csv('social-network-ads.csv')
ads
```

```python
X = ads.iloc[:, :-1].values
y = ads.iloc[:, -1].values
```

```python
X
```

```python
y
```

### Data Splitting & Preprocessing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Model Training & Predicting

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train) #hna x train w ytrain 3ks abl kda kan x train bs
```

```python
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

### Model Evaluation

```python
from sklearn.metrics import confusion_matrix, accuracy_score , plot
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

### Choosing Best Value of 'K'

```python
test_error_rates = []

for k in range(1,30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(X_train,y_train) 
   
    y_pred_test = knn_model.predict(X_test)
    
    test_error = 1 - accuracy_score(y_test,y_pred_test)
    test_error_rates.append(test_error)
```

```python
test_error_rates
```

```python
plt.figure(figsize=(10,6),dpi=200)
plt.plot(range(1,30),test_error_rates,label='Test Error')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel("K Value")
```
